```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

window_sum = sum(arr[:k])
max_sum = window_sum

for i in range(k, len(arr)):
    current_number = arr[i]
    first_element_of_window = arr[i-k]
    window_sum += current_number - first_element_of_window

    if window_sum > max_sum:
        max_sum = window_sum

print(max_sum)
```

```python
#               |
s = "cxcdfsaqwebcsdfdsabcdabcdefghy"
#     ^


seen = set()
left = 0
max_len = 0

for right, ch in enumerate(s):
    while ch in seen:
        seen.remove(s[left])
        left += 1
    seen.add(ch)

    curent_window_width = right - left + 1
    max_len = max(max_len, curent_window_width)

print(max_len)
```
