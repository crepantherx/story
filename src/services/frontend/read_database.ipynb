```python
import chromadb
import pandas as pd

client = chromadb.PersistentClient(path="./data/chroma_db_final")
col = client.get_collection("profiles")

# Fetch everything in chunks (Chroma caps results per call)
all_data = []
limit = 5000
offset = 0

while True:
    batch = col.get(offset=offset, limit=limit)
    ids = batch.get("ids", [])
    metas = batch.get("metadatas", [])
    if not ids:
        break
    for id_, meta in zip(ids, metas):
        all_data.append({"id": id_, **(meta or {})})
    offset += limit

df = pd.DataFrame(all_data)
print("✅ Reconstructed DataFrame with", len(df), "rows.")
df.head()
```
